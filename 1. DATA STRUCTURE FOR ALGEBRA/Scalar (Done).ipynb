# **Soal no. 2**
Lakukan tahapan pada soal nomor 1 dengan menggunakan TENSORFLOW

```python
import tensorflow as tf
```

```python
# Buat variabel skalar x dan simpan suatu nilai skalar
# Tampilkan tipe data dari x
x_tf = tf.Variable(110, dtype=tf.int16)
x_tf
```

```python
# Buat variabel skalar y dan simpan suatu nilai skalar
# Tampilkan tipe data sari y
y_tf = tf.Variable(2515, dtype=tf.int16)
y_tf
```

```python
# Jumlahkan x+y
tf_sum = tf.add(x_tf, y_tf)
tf_sum
```

```python
# Buat variabel bernama x_float dan simpan suatu nilai desimal
# Tampilkan tipe data x_float
x_float_tf = tf.Variable(61.110, dtype=tf.float16)
x_float_tf
```

```python
# Jumlahkan x_float dengan y
tf_sum = x_float_tf + tf.cast(y_tf, dtype=tf.float16)
tf_sum
```

# **Soal no. 3**
Lakukan tahapan pada soal 1 dengan menggunakan PYTORCH

```python
import torch as pt
```

```python
# Buat variabel skalar x dan simpan suatu nilai skalar
# Tampilkan tipe data dari x
x_pt = pt.tensor(110, dtype = pt.int16)
x_pt
```

```python
# Buat variabel skalar y dan simpan suatu nilai skalar
# Tampilkan tipe data dari y
y_pt = pt.tensor(2515, dtype = pt.int16)
y_pt
```

```python
# Jumlahkan x+y
pt_sum = pt.add(x_pt, y_pt)
pt_sum
```

```python
# Buat variabel bernama x_float dan simpan suatu nilai desimal
# Tampilkan tipe data x_float
x_float_pt = pt.tensor(61.110, dtype=pt.float16)
x_float_pt
```

```python
# Jumlahkan x_float dengan y
pt_sum = x_float_pt + y_pt
pt_sum
```
