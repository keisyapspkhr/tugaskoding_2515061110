# **Soal No 6**
Buktikan apakah perkalian matriks berdifat komutatif?

(AB=BA)?

```python
np.dot(A, B)
```

```python
np.dot(B, A)
```

Aturan perkalian matriks,

Dua matriks bisa dikalikan jika:

**jumlah KOLOM matriks PERTAMA = jumlah BARIS matriks KEDUA**

**(mxn)(nxp)**

Sedangkan pada soal,

B=(3x3) A=(2x3)

maka (3x3)(2x3)

kolom matriks pertama=3

baris matriks kedua=2

maka tidak bisa dikalikan dan seperti kode diatas akan terjadi error.

**Kesimpulan:**

perkalian matriks **TIDAK bersifat komutatif**.
