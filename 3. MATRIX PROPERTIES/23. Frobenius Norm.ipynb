# **Soal No 1**
Buatlah matrix ukuran 3x3 lalu hitung nilai Frobenius Norm

NUMPY

```python
import numpy as np
```

```python
X = np.array([[1, 1, 0], [2, 3, 9], [2, 5, 6]])
X
```

```python
np.linalg.norm(X)
```

PYTORCH

```python
import torch
```

```python
X_pt = torch.tensor([[11, 61, 15], [23, 9, 6], [21, 4, 26]])
X_pt
```

```python
torch.norm(X_pt.float())
```

TENSORFLOW

```python
import tensorflow as tf
```

```python
X_tf = tf.Variable([[18, 30, 1], [3, 34, 28.], [2, 5, 7]])
X_tf
```

```python
tf.norm(X_tf)
```
