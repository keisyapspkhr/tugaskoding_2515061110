```python
import numpy as np
```

```python
# Buat matrix X ukuran 2x2
X = np.array([[2, 5],
              [6, 1]])
X
```

```python
# Buat invers dari matrix X dan beri nama Xinv
Xinv = np.linalg.inv(X)
Xinv
```

```python
# Kalikan X dengan Xinv
np.dot(Xinv, X)
```

**NUMPY**

```python
import numpy as np
```

```python
V = np.array([[4, 2],
              [-5, -3]])
V
```

```python
W = np.array([[4],
              [-7]])
W
```

```python
Vinv = np.linalg.inv(V)
Vinv
```

```python
X = np.matmul(Vinv, W)
X
```

**PYTORCH**

```python
import torch
```

```python
V_pt = torch.tensor([[4, 2],
                     [-5, -3]])
V_pt
```

```python
W_pt = torch.tensor([[4],
              [-7]])
W_pt
```

```python
Vinv_pt = torch.inverse(V_pt.float())
Vinv_pt
```

```python
X_pt = torch.matmul(Vinv_pt, W_pt.float())
X_pt
```

**TENSORFLOW**

```python
import tensorflow as tf
```

```python
V_tf = tf.Variable([[4, 2],
                     [-5, -3]])
V_tf
```

```python
W_tf = tf.Variable([[4],
              [-7]])
W_tf
```

```python
Vinv_tf = tf.linalg.inv(tf.cast(V_tf, dtype=tf.float32))
Vinv_tf
```

```python
X_tf = tf.matmul(Vinv_tf, tf.cast(W_tf, dtype=tf.float32))
X_tf
```
