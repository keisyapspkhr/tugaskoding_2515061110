# **Soal No 1**

```python
import numpy as np
```

Buat vector i dengan elemen (1,0) dan vector j dengan elemen (0,1)

```python
# vector i dengan elemen (1,0)
i = np.array([1, 0])
i
```

```python
# Vector j dengan elemen (0,1)
j = np. array([0,1])
j
```

```python
# Buktikan bahwa vector i dan vector j itu orthogonal
np.dot(i, j)
```
