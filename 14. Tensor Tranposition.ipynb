```python
import numpy as np
import torch
import tensorflow as tf
```

# **Soal No 1**

```python
# Buat matrix X dengan ukuran 2x3 menggunakan NUMPY
X = np.array([[23, 9, 6], [25, 61, 110]])
X
```

```python
# Buat matrix X_pt dengan ukuran 2x3 menggunakan PYTORCH
X_pt = torch.tensor([[6, 9, 23], [110, 61, 25]])
X_pt
```

```python
# Buat matrix X_tf dengan ukuran 2x3 menggunakan TENSORFLOW
X_tf = tf.Variable([[9, 6, 23], [61, 15, 25]])
X_tf
```
