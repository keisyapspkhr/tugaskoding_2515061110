# **Soal No 3**


```python
import torch
import tensorflow as tf
```

Soal nomor 2 menggunakan PYTORCH

```python
x_pt = torch.tensor([[25],
                     [15],
                     [110]])
x_pt
```

```python
# Lakukan tranpose pada vektor x dan simpan pada variabel x_t
x_pt_t = x_pt.T
x_pt_t
```

```python
# Tampilkan ukuran dari vektor x_t
x_pt.shape
```

```python
y_pt = torch.tensor([[25, 61, 110]])
y_pt
```

```python
# Lakukan transpose pada vector y dan simpan pada variabel y_t
y_pt_t = y_pt.T
y_pt_t
```

```python
# Tampilkan ukuran dari vector x_t
y_pt.shape
```

Soal nomor 2 menggunakan TENSORFLOW

```python
x_tf = tf.Variable([[25],
                     [15],
                     [110]])
x_tf
```

```python
# Lakukan tranpose pada vektor x dan simpan pada variabel x_t
# Tampilkan ukuran dari vektor x_t
x_tf_t = tf.transpose(x_tf)
x_tf_t
```

```python
y_tf = torch.tensor([[25, 61, 110]])
y_tf
```

```python
# Lakukan transpose pada vector y dan simpan pada variabel y_t
# Tampilkan ukuran dari vector x_t
y_tf_t = tf.transpose(y_tf)
y_tf_t
```
