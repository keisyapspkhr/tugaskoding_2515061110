# **Soal No 1**
Buat visualisasi persamaan berikut

y = 2x+1 dan 3x-y = 5

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000)
```

```python
y1 = 2*x + 1
```

```python
y2 = 3*x - 5
```

```python
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
ax.set_xlim([0, 10])
ax.set_ylim([0, 15])
ax.plot(x, y1, c='green')
ax.plot(x, y2, c='brown')
plt.axvline(x=6, color='purple', linestyle='--')
_ = plt.axhline(y=13, color='purple', linestyle='--')
```
