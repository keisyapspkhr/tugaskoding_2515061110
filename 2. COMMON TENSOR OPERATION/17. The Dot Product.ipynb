# **Soal No 1**

Buat vector x dan y dengan ukuran 1x3 menggunakan NUMPY

```python
import numpy as np
```

```python
x = np.array([2, 1, 5])
x
```

```python
y = np.array([6, 1, 11])
y
```

Buat vector x_pt dan y_pt dengan ukuran 1x3 menggunakan PYTORCH

```python
import torch
```

```python
x_pt = torch.tensor([25, 15, 61])
x_pt
```

```python
y_pt = torch.tensor([23, 9, 6])
y_pt
```

Buat vector x_tf dan y_tf dengan ukuran 1x3 menggunakan TENSORFLOW

```python
import tensorflow as tf
```

```python
x_tf = tf.Variable([15, 61, 11])
x_tf
```

```python
y_tf = tf.Variable([20, 6, 23])
y_tf
```

# **Soal No 2**

```python
# Hitung perkalian DOT PRODUCT antara vektor x dan y
np.dot(x, y)
```

```python
# Hitung perkalian DOT PRODUCT antara vektor x_pt dan y_pt
torch.dot(x_pt, y_pt)
```

```python
# Hitung perkalian DOT PRODUCT antara cektor x_tf dan y_tf
tf.reduce_sum(tf.multiply(x_tf, y_tf))
```
